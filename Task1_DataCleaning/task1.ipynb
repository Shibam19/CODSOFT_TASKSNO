```python
print("Summary Statistics (Numerical Columns):")
df.describe()
```

```python
print("Summary Statistics (Categorical Columns):")
df.describe(include='object')
```

```python
print("Missing Values Before Cleaning:")
print(df.isnull().sum())
```

```python
missing = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Percentage": (df.isnull().sum()/len(df))*100
})

print(missing)
```

```python
print("Duplicate Rows Before Cleaning:")
print(df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)

print("Duplicate Rows After Cleaning:")
print(df.duplicated().sum())
```

```python
print("Data Types:")
print(df.dtypes)
```

```python
df["OrderDate"] = pd.to_datetime(df["OrderDate"])
```

```python
categorical_columns = df.select_dtypes(include='object').columns

for col in categorical_columns:
    print(f"\nUnique values in {col}:")
    print(df[col].unique())
```

```python
for col in categorical_columns:
    df[col] = df[col].str.strip()
```

```python
for col in categorical_columns:
    df[col] = df[col].str.title()
```

```python
numeric_columns = df.select_dtypes(include=np.number).columns

for col in numeric_columns:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].median())
```

```python
for col in categorical_columns:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
print("Missing Values After Cleaning:")
print(df.isnull().sum())
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_")
```

```python
print(df.columns)
```

```python
print("Final Dataset Information:")
df.info()
```

```python
df.to_csv("Cleaned_Regional_Sales.csv", index=False)

print("Cleaned dataset saved successfully as CSV.")
```
